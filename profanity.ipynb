```python
from profanity_check import predict, predict_prob
from langdetect import detect
```

```python
import glob
file_list = glob.glob("List-of-Dirty-Naughty-Obscene-and-Otherwise-Bad-Words-master/*")
```

```python
all_tokens = []
for file in file_list:
    with open(file, encoding="utf-8") as f:
        content = f.readlines()
    content = [x.strip() for x in content]
    all_tokens.append(content)
    
list_of_tokens = set([item for sublist in all_tokens for item in sublist])
```

```python
text_input = 'you'

if(detect(text_input) == "en"):
    print(predict_prob([text_input.lower()]))
else:
    for l in list_of_tokens:
        if(l in text_input.lower()):
            print(l, [0.99999])
            break;
```
